**Menginstal Library**

```python
#Pertama-tama kita import beberapa library ke dalam program ini seperti pandas, numpy, dan apyori
!pip install apyori

import pandas as pd
import numpy as np

from apyori import apriori
```

**Meload Dataset**

```python
# Meload dataset yang berisikan ID Transaksi dan Items 
df= pd.read_excel('/content/drive/MyDrive/Items.xlsx')
```

**Cleaning Data**

```python
# Menghapus kolom ID Transaksi
dropped = ['ID Transaksi']
df.drop(columns=dropped, inplace=True)
df.head()
```

```python
# Melihat shape dari tabel, yang sebelumnya 2 kolom 11 baris
df.shape
```

```python
# Membuat variabel data
data = df
```

**Membuat Data Training**

```python
records = []
for i in range(data.shape[0]):
    records.append([str(data.values[i,j]).split(',') for j in range(data.shape[1])])

trx = [[] for trx in range(len(records))]
for i in range(len(records)):
    for j in records[i][0]:
        trx[i].append(j)
trx
```

**Melakukan Perhitungan Apriori**

```python
# Menggunakan fungsi apriori untuk membuat asosiasi
association_rules = apriori (trx)
# Membuat list untuk menampilkan hasil
association_results = association_rules
```

**Menampilkan Hasil Perhitungan**

```python
# Menampilkan hasil dari penerapan assosiation rule
pd.set_option('max_colwidth', 1000)
Result=pd.DataFrame(columns=['Rule','Support','Confidence'])
for item in association_results:
    pair = item[2]
    for i in pair:
        items = str([x for x in i[0]])
        if i[3]!=1:
            Result=Result.append({
                'Rule':str([x for x in i[0]])+ " -> " +str([x for x in i[1]]),
                'Support':str(round(item[1]*100,2))+'%',
                'Confidence':str(round(i[2] *100,2))+'%'
                },ignore_index=True)
Result
```
